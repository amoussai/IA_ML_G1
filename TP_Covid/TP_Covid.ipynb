```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sp-ti-tp-7j-reg-2021-04-12-18h12.csv", sep=';')
df
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['nom_region'].nunique()
```

```python
df['nom_region'].unique()
```

```python
def plot_covid_evol(week, pos):
    plt.figure(figsize=(20, 8))
    plt.plot(week, pos )
    plt.xticks(rotation=90, ha='right')
    plt.show()
```

```python
df_idf = df.loc[df['nom_region'] == 'Ile-de-France']
df_idf
```

```python
plot_covid_evol(week=df_idf.semaine_glissante[:100], pos=df_idf.P[:100])
```

```python
df_corse = df.loc[df['nom_region'] == "Provence-Alpes-Côte d'Azur"]
df_corse
```

```python
plot_covid_evol(week=df_corse.semaine_glissante[:100], pos=df_corse.P[:100])
```

```python
plt.figure(figsize=(20, 8))
plt.plot(df_corse.semaine_glissante[:100], df_corse.P[:100] )
plt.plot(df_idf.semaine_glissante[:100], df_idf.P[:100] )
plt.xticks(rotation=90, ha='right')
plt.show()
```
