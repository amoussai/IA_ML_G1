#### méthode apply()

```python
# series
s_superficie = df["Superficie"]
s_superficie
```

```python
# côtés des carrés de surface équivalente
s_superficie.apply(np.sqrt)
```

```python
df["Superficie"].apply(np.sqrt)
```

```python
# rayons des cercles de surface équivalente
# fonction standard
def rayon(surface):
    rayon = np.sqrt(surface / np.pi)
    return rayon

rayon(10)
```

```python
# rayons des cercles de surface équivalente
# fonction standard
df['Superficie'].apply(rayon)
```

```python
y =  lambda x : rayon(x)
df['Superficie'].apply(lambda x : rayon(x))
```

```python
# rayons des cercles de surface équivalente
# lambda
s_superficie.apply(lambda x: np.sqrt(x / np.pi))
```
